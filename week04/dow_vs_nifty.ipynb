```python
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
nifty_dow_data = yf.download(['^NSEI', '^DJI'])['Adj Close']
nifty_dow_data = nifty_dow_data.fillna(method='ffill')
nifty_dow_data.dropna(inplace=True)
nifty_dow_data.plot()
plt.show()
```

```python
nifty_dow_data.pct_change().add(1).cumprod().plot()
plt.show()
```

```python
returns_df = nifty_dow_data.pct_change().copy()
plt.scatter(returns_df['^NSEI'], returns_df['^DJI'])
plt.show()

returns_df.corr()
```
